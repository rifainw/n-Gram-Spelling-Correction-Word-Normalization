```python
# Import library
import pandas as pd
import re
import string
import nltk
```

```python
# Download required library
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
# Loading the dataset
df = pd.read_csv(r'C:\Users\ASUS\Downloads\dl_101-master\data\spam.csv', delimiter = ',', encoding = 'latin-1')
```

```python
df = df.drop(df.columns[[2, 3, 4]], axis=1)
```

```python
df.columns = ['Label','Message']
df.head()
```

```python
# Text Cleaning
def clean_txt(txt):
    txt = "".join([c for c in txt if c not in string.punctuation])
    tokens = re.split('\W+', txt)
    txt = " ".join([ps.stem(word) for word in tokens if word not in stopwords])
    return txt
```

```python
df[ 'Message_Clean' ] = df['Message'].apply(lambda x: clean_txt(x))
```
